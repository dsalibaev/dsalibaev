```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загружаем готовый датасет из интернета
# (Я подобрал для тебя прямую ссылку на данные о треках)
# Корректная ссылка на CSV-файл со списком песен Spotify
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv'
df = pd.read_csv(url)

# Посмотрим на первые 5 строк
df.head()
```

```python
# Строим график рассеяния (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='energy', y='danceability')
plt.title('Связь между Энергией и Танцевальностью песен')
plt.grid(True)
plt.show()
```
